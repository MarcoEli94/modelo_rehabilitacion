## 1. Introducción

El sistema *ModeloRehabilitacion* es una solución completa de rehabilitación motriz basada en inteligencia artificial, desarrollada con Python. El sistema integra modelos de aprendizaje profundo (TCN - Temporal Convolutional Networks) para la clasificación de movimientos correctos/incorrectos, una API REST moderna para la exposición de servicios, y herramientas de análisis de datos para el procesamiento de secuencias biomecánicas. Este documento describe el stack tecnológico, las dependencias principales, la arquitectura de software y las referencias de las tecnologías utilizadas.

## 5. Pipeline de Aprendizaje Automático

### 5.1. Arquitectura del Modelo TCN

Los modelos TCN (Temporal Convolutional Networks) implementan:
- **Capas convolucionales causales** para procesamiento temporal
- **Dilatación exponencial** (1, 2, 4, 8) para campo receptivo amplio
- **Bloques residuales** con conexiones skip
- **Regularización** (dropout 0.15) para evitar sobreajuste
- **Optimización** con Adam y early stopping

### 5.2. Features Biomecánicos

Cada secuencia de 48 frames incluye 10 features derivadas:
1. Ángulo de hombro derecho
2. Relación muñeca/codo normalizada
3. Relación codo/hombro normalizada
4. Elevación de hombro
5. Inclinación del tronco
6. Diferencia de simetría
7. Disponibilidad de simetría
8. Velocidad angular
9. Progreso de rango de movimiento
10. Ratio de puntos válidos

### 5.3. Entrenamiento y Validación

- **Dataset:** UI-PRMD (movimientos m07/m09)
- **División:** 80% entrenamiento, 20% validación
- **Métricas:** Recall >90%, F1-score, precisión
- **Balanceo:** Técnicas SMOTE para clases desbalanceadas
- **Persistencia:** Modelos guardados en formato .keras + escaladores .joblib

## 8. Decisiones Tecnológicas Clave

### 8.1. ¿Por qué FastAPI en lugar de Flask/Django?

1. **Rendimiento ASGI:** Superior a WSGI para APIs modernas
2. **Validación automática:** Pydantic integrado
3. **Documentación automática:** OpenAPI/Swagger sin configuración
4. **Tipado fuerte:** Mejor soporte IDE y debugging
5. **Async nativo:** Operaciones no bloqueantes eficientes

### 8.2. ¿Por qué TensorFlow en lugar de PyTorch?

1. **Madurez en producción:** Mejor soporte para despliegue
2. **Keras API:** Simplicidad para prototipado rápido
3. **Optimizaciones:** Mejor rendimiento en CPU para inferencia
4. **Ecosistema:** Amplio soporte de herramientas
5. **Compatibilidad:** Mejor integración con scikit-learn

### 8.3. ¿Por qué TCN en lugar de LSTM/Transformers?

- **Eficiencia temporal:** Convoluciones más rápidas que RNN
- **Paralelización:** Procesamiento batch eficiente
- **Interpretabilidad:** Estructura más simple que Transformers
- **Requisitos de datos:** Menos datos que Transformers para converger
